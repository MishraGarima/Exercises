### The Data
Working with titanic data set.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

### Exercises

**Recreate the plots below using the titanic dataframe. Keep careful attention to the x and y labels for hints.**

```python
sns.jointplot(x='fare', y='age', data=titanic, color='lime')
```

```python
sns.distplot(titanic['fare'], bins=30, kde=False, color='Red')
```

```python
sns.boxplot(x=titanic['class'], y=titanic['age'], palette='Set1')
```

```python
sns.swarmplot(x='class', y='age', data=titanic, palette='Accent')
```

```python
sns.countplot(titanic['sex'], palette='viridis')
```

```python
sns.heatmap(titanic.corr(),cmap='RdPu')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titanic, col='sex')
g.map(plt.hist,'age')
```

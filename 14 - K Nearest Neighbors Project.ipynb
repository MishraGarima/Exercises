# K Nearest Neighbors Project 

### Import Libraries and Get the Data
**Import Libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

**Get the Data**

```python
df = pd.read_csv('KNN_Project_Data')
```

**Check the head of the dataframe.**

```python
df.head()
```

### EDA

**Use seaborn on the dataframe to create a pairplot with the hue indicated by the TARGET CLASS column.**

```python
sns.set()
sns.pairplot(df, hue='TARGET CLASS', corner=True, palette='seismic')
```

### Standardize the Variables
**Import StandardScaler from Scikit learn**

```python
from sklearn.preprocessing import StandardScaler
```

**Create a StandardScaler() object**

```python
scaler = StandardScaler()
```

**Fit scaler to the features**

```python
scaler.fit(df.drop('TARGET CLASS',axis=1))
```

**Use the .transform() method to transform the features to a scaled version**

```python
scaled_features = scaler.transform(df.drop('TARGET CLASS',axis=1))
```

**Convert the scaled features to a dataframe and check the head of this dataframe to make sure the scaling worked**

```python
scaled_data = pd.DataFrame(scaled_features, columns=df.columns[:-1])
```

```python
scaled_data.head()
```

### Train Test Split
**Use train_test_split to split your data into a training set and a testing set**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_features,df['TARGET CLASS'], test_size=0.3, 
                                                    stratify=df['TARGET CLASS'])
```

### Using KNN

**Import KNeighborsClassifier from scikit learn.**

```python
from sklearn.neighbors import KNeighborsClassifier
```

**Create a KNN model instance**

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

**Fit this KNN model to the training data**

```python
knn.fit(X_train,y_train)
```

### Predictions and Evaluations

**Predict values using your KNN model**

```python
pred = knn.predict(X_test)
```

**Create a confusion matrix and classification report**

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print("Confusion Matrix : \n", confusion_matrix(y_test,pred))
print("Classification Report : \n", classification_report(y_test,pred))
```

### Choosing a K Value

**Create a for loop that trains various KNN models with different k values, then keep track of the error_rate for each of these models with a list**

```python
error_rate = []

for i in range(1,50):    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

**Create the plot using the error rate**

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,50),error_rate,color='goldenrod', linestyle='dashed', marker='o', markerfacecolor='firebrick',
         markersize=7)
plt.title('K-Value vs Error Rate')
plt.xlabel('K value')
plt.ylabel('Error Rate')
```

### Retrain with new K Value

**Retrain your model with the best K value and re-do the classification report and the evaluation**

```python
knn = KNeighborsClassifier(n_neighbors=44)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
print("K = 44")
print("Confusion Matrix : \n", confusion_matrix(y_test,pred))
print("Classification Report : \n", classification_report(y_test,pred))
```
